**Write a Python program that takes a list of numbers and
counts how many even numbers are in the list.**

```python
lis = [48, 42, 6, 72, 42, 64, 44, 62, 54, 33, 51, 33, 32, 64, 12, 33, 19, 82, 9, 44]

def count_even(lis):
  counts = 0
  for i in lis:
    if i % 2 == 0:
      counts += 1
  return counts

count_even(lis)
```
