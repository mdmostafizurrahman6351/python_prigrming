Write a Python program that takes a list of numbers and
calculates the average (mean) of the numbers.

```python

def average(lst):
    total = 0
    count = 0
    for x in lst:
        total += x
        count += 1
    if count == 0:
        return None   # খালি তালিকার জন্য None রিটার্ন করা হলো
    return total / count

lis = [1, 2, 45, 55, 44, 66, 66, 77]
print(average(lis))
```

Write a Python program that takes a list of numbers and
creates a new list containing only odd numbers.

```python

def odd_list(lis):
  new_odd_list = []
  for i in lis:
    if i % 2 != 0:
      new_odd_list.append(i)

  return new_odd_list

lis = [1,2,45,55,44,66,66,77]
odd_list(lis)
```

Write a Python program that takes a list of numbers and
finds the difference between the largest and smallest number in the list.

```python
def difference(lis):
    smallest = lis[0]
    largest = lis[0]
    for i in lis:
        if i > largest:
            largest = i
        if i < smallest:
            smallest = i
    betwen_difference = largest - smallest
    return betwen_difference

lis = [1,2,4,5,5,66,66,55,66,44,5,44,90,76,65]
difference(lis)
```

Write a Python program that takes a list of numbers and
calculates the frequency of each number.
Then sort the numbers by frequency in descending order and
return the Top 3 items as (number, count).
If two numbers have the same frequency, sort them by
number in ascending order.

```python

def top_3(lis):
  number_count = {}

  for i in lis:
   if i in number_count:
     number_count[i] += 1
   else:
      number_count[i] = 1

  items = []
  for k, v in number_count.items():
    items.append((k, v))

  sorted_items = sorted(items, key=lambda x: (-x[1], x[0]))

  top_3 = sorted_items[:3]
  return top_3
top_3([3, 1, 2, 3, 4, 2, 3, 1])
```

Write a Python program that takes a list of numbers and
calculates the factorial of each number.
Factorial of 0 = 1
Use Python basics and loops, do not use math.factorial()

```python

def list_factorial(lis):
  fac_list = []

  for i in lis:
    if i in (0, 1):
      fac_list.append(1)
      continue
    fac = 1
    for j in range(2, i+1):
      fac *= j
    fac_list.append(fac)
  return fac_list
list_factorial([0,1,2,3])
```

Write a Python program that takes a list of numbers and
marks whether each number is prime or not.
Output should be a list of True/False
1 is not prime
Use Python basics and loops, do not use any built-in prime functions.

```python

def prime_check(lis):
  prime_list = []
  for i in lis:
    if i < 2:
      prime_list.append(False)
      continue
    is_prime = True
    for j in range(2, int(i ** 0.5) + 1):
      if i % j == 0:
        is_prime = False
        break
    prime_list.append(is_prime)
  return prime_list

prime_check([1,2,4,56,44,77,41])
```

Write a Python program that takes a list of numbers and
removes duplicates from the list.
Output should contain only unique numbers.
Use Python basics and loops, do not use set().

```python

def removed_duplicates(lis):
  unique_list = []

  for i in lis:
    if i not in unique_list:
      unique_list.append(i)
  return unique_list
removed_duplicates([1,2,4,6,4,3,2,1,7])
```

Write a Python program that takes a list of numbers and
finds the second largest number in the list.

Condition:
Python basic only using

max() and sorted() don't use

```python

def second_largest(lis):

  largest = lis[0]
  for i in lis:
    if i > largest:
      largest = i

  second_large = None
  for x in lis:
    if x != largest:
      if second_large is None or x > second_large:
        second_large = x
  return second_large

second_largest([1,2,3,4,4,5,6,6,8])
```

Write a Python program that takes a list of numbers and
returns the cumulative sum while preserving order.

```python

def cumulative_sum(lis):
  cum_list = []
  total = 0
  for i in lis:
    total = i + total
    cum_list.append(total)
  return cum_list

cumulative_sum([1,2,3,5,5,6,8])
```

Write a Python program that takes a list of numbers and a given target,
and finds all unique pairs whose sum equals the target.

```python

def fill_target(lis, target):
  match_list = []
  for i in range(len(lis)):
    for j in range(i+1, len(lis)):
      if lis[i] + lis[j] == target:
        match_list.append((lis[i], lis[j]))
  return match_list
fill_target([1,2,3,4,-4,-5,6,0,-2,5], 3)
```

Transform the English letters in a string: convert all uppercase letters to lowercase and all lowercase letters to uppercase.
Example: string = "Hello World" → output = "hELLO wORLD"

```python

def string(text_data):
  text_data = text_data.title()
  text_data = text_data.swapcase()

  return text_data
text_data = "Hello. How are you?"
string(text_data)
```

Write a program to count how many vowels (a, e, i, o, u — both uppercase and lowercase) are in an English string.
Example: "Data Science" → output = 5

```python

text = "Data science is fun."
vowels = "aeiou"
def count_vowels(text):
  text = text.lower()
  count = 0
  for i in text:
    if i in vowels:
      count += 1
  return count
count_vowels(text)
```
