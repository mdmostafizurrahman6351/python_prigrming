Given a list of integers, return a new list containing only the even numbers, preserving their original order.

```python

def get_even_numbers(nums):
    return [num for num in nums if num % 2 == 0]

get_even_numbers([1,2,3,4,5,5,6,6,76])
```

Given a string, return a dictionary with the count of each character (case-sensitive).

```python

def char_count(s):
  char_dic = {}
  for char in s:
    if char in char_dic:
      char_dic[char] += 1
    else:
      char_dic[char] = 1
  return char_dic
char_count("datascience")
```

```python

def char_count(s):
  char_dic = {}
  for char in s:
    char_dic[char] = char_dic.get(char, 0) + 1
  return char_dic
char_count("datascience")
```

```python

from collections import Counter

def char_count(s):
    return dict(Counter(s))
char_count("datascience")
```

Given a sentence (string), return the longest word in the sentence.
If there are multiple words with the same maximum length, return the first one.

```python

def longest_word(sentence):
  if len(sentence) < 1:
    return None
  sentence = sentence.split()
  largest_len = "0"
  for word in sentence:
    if len(word) > len(largest_len):
      largest_len = word
  return largest_len

longest_word("python is the best language in the serbises")
```

```python

def longest_word(sentence):
    word = sentence.split()
    return max(word, key=len) if word else None

longest_word("python is the best language in the services")
```
